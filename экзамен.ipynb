Напишите функции в системе MatLab, которые позволяют выполнить дискретное вейвлет-преобразование одного уровня. Описать входные и выходные переменные
а) с вейвлетом Мейера  meyr
б) с вейвлетом Хаара   haar
в) вейвлетом Добеши 8 порядка db8

```python
import numpy as np
import pywt

from pylab import rcParams

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# генерируем данные

n = 2048

a = np.arange(n)
noise = np.random.normal(0,1,n)
signal = np.sin(a) + noise
```

```python
rcParams['figure.figsize'] = 20, 10
plt.plot(signal)
```

```python
def plot(cA, cD,itr=None):
    f, axarr = plt.subplots(2, sharex=True)
    axarr[0].plot(cA)
    if itr:
        axarr[0].set_title('Iteration ' + str(itr))
    axarr[1].plot(cD, c="red")
```

```python
meyer = pywt.Wavelet('dmey')
haar = pywt.Wavelet('haar')
db8 = pywt.Wavelet('db8')
```

```python
cA, cD = pywt.dwt(signal, meyer) # Approximation and detail coefficients.

plot(cA,cD)
```

```python
cA, cD = pywt.dwt(signal, haar) # Approximation and detail coefficients.

plot(cA,cD)
```

```python
cA, cD = pywt.dwt(signal, db8) # Approximation and detail coefficients.

plot(cA,cD)
```
